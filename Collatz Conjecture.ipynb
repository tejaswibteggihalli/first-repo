```python
def CollatzConjecture(n):
    step = 0
    m = n
    while not n == 1:
        k = n
        # n is even 
        if n % 2 == 0:
            n /= 2
        else:
            # n is odd
            n *= 3
            n += 1
        step += 1
        print(f"Step:{step}\nNumber changed from {k} to {n}")
    print(f"Hence Collatz Conjecture has held true!\nNumber {m} was transformed into 1 in {step} steps!")
        
            
        
            
```

```python
CollatzConjecture(998)
```
